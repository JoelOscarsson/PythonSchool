## Example - birds and oil
There is a oil leakage causing the bird population in an island to 1/2 in each year. From start there were 80000 birds, how many years does it take for it to have 1/10 remanining?

````python
birds, year = 80000, 0 #tuple unpacking kallas det då man kan skriva kommatecken och fortsätta liksom

while birds >= 8000:
    birds = birds/2
    year += 1
print(f"Years={year}")
````

## Example - Milk

- 1L of milk has 1 500 000 bacterias in room temperature
- Bacrerias increases with 50% each hour in room temp
- Milk get soured after 10 000 000 bacterias
- How many hours does it take til milk sours (whole hours)

````python
bacteries = 1500000
hour = 0

while bacteries <= 10000000:
    bacteries *= 1.5
    hour += 1

print(f"The milk sours after this many hours={hour}")
````

````python
##Facit för uppgiften över
import matplotlib.pyplot as plt

bacterias = 1.5e6 # 1.5 * 10^6   ##exponent(då har man 10 i) står för upphöjt
factor = 1.5
hours = 0 
sour = 1e7 # 10^7

bacterias_list = [bacterias]

print(bacterias_list)

while bacterias < sour:
    bacterias *= factor
    hours += 1
    bacterias_list.append(bacterias)
print(bacterias_list)
print(f"Takes {hours}h for milk to get soured")

# creates a range object with numbers: 0,1,2...,hours
hours_list = range(0, hours+1)

plt.plot(hours_list, bacterias_list, 'o--')
plt.plot(hours_list, [sour]*(hours+1))
plt.grid()
plt.xlabel("hours")
plt.ylabel("number of bacterias")
plt.title("bacerias in milk room temperature")
````
