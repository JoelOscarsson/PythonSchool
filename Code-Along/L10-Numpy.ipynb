## Numpy

```python


numbers_list = [1,2,3]   # En vanlig lista
numbers_list = numbers_list*2 #Kopierar sig självt och läggs in i en lista
print(numbers_list)

# in a list we can place different data types
mixed_list = [1, "ett", True, 2.3]   ## En lista, samt lite olika kommands. ## Dynanamic typin in python 
# One element in the list is an obejt, it has information on what type it is.
# This requires some overhead, which slows down the program. 
# Vissa hävdar att python är långsamt, bara för det är dynamiskt typat (just a sidenote)
# Note that when we have one data type, it is unnecessary to have this overheaad and thats why numpy is good.
print(mixed_list)  
```

```python
# in numpy we have statically typed
import numpy as np

# Create an array from a list
vector1 = np.array([2,5,1])
print(vector1)

## Elementvise multiplication
vector2 = vector1*2   ## Ggr med vector1
print(vector2)

## Indexing
print(f"{vector1[2]=}")  
print(f"{vector1[-1]=}")  

# Slicing
print(f"{vector1[1:]=}")  


vector1[1] = 1337    ## Notera vad som hänt med indexen eftersom du hämtat 1:an
print(vector1)

# Can't append(lägga till) an array - can't change it's size. This is something u can do in a list but not an array.
# Can't have different types in an array.

# Sort
vector1.sort()
print(f"{vector1=}")

## Vi kan beräkna summa
print(f"{vector1.sum()=}")   ## Det är mycket snabbare att använda numpy, pga vektorisering. Själva loopandet pushas till C
# Man ska inte loopa när man jobbar med numpy

print(f"{vector1.sum()=:.2f}")   # Man kan formatera





```

```python
ones_vector = np.ones(10)
print(f"{ones_vector=}")


# Skapar en 2d array of shape (3,3)  En matrix 3x3
matrix_3x3 = np.ones((3,3))
print(f"{matrix_3x3=}") 

# [Row, index] - counts from 0
matrix_3x3[2,1] = 10
matrix_3x3
```

```python
## Man kan köra 0:or

np.zeros((2,5))     ## notera parantesen. - Shapelike - tar in en shape,en lista eller en tuple

```

```python
print(matrix_3x3.shape)

## Slicing operator to get a row
print(f"{matrix_3x3[2,:]=}")  ## Vill komma åt sista raden

# Slicing operator to get a column
print(f"{matrix_3x3[:,1]=}") # Slicing operator to get a column

## Vi kan add 2 vectors
matrix_3x3[2,:] + matrix_3x3[:,1]
```

## Linspace
- Skapar en vektor

```python
x = np.linspace(-5,5)  # Ger oss per default 50 equally spaced numbers between low and high
print(x.shape)
```

```python
import matplotlib.pyplot as plt 

#  Samma sak som att skriva <=> def f(x): return x + 2
f = lambda x: x+2
g = lambda x: -x 

print(f(x))    # Plottar x mot f x
plt.plot(x, f(x))   
plt.plot(x, g(x))
```
