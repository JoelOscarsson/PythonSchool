## Matplotlib

```python
import matplotlib.pyplot as plt  # Betyder att plt existerar i vårt namespace
import numpy as np # np ligger i vårt namespace


# array from -10 to 10 with 100 numbers equally spaced between
x = np.linspace(-10,10, 100)


f = lambda x: np.cos(x)
# lambda är lika med nedanför
# def f(x):
#   return np.cos(x)
```

## Matlab style interface

```python
plt.plot(x, f(x), 'o--')
```

```python
x = np.linspace(-10,10, 100)

plt.figure(figsize=(8, 4), dpi = 200)  #ger bredare width på vågorna  #dpi = dots per inches
plt.plot(x, f(x))
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
#plt.box(False)
```

## OOP approach

```python
# Instanserat fig, ax from Figure and axes
fig, ax = plt.figure(figsize = (8,4)), plt.axes()
ax.plot(x, f(x))  #ax object has a plot method
ax.set(xlabel =  "x", ylabel="y", title = "$\cos{x}$", xlim = (-5.5), ylim = (-1.2,1.2))
fig.savefig("cos(x).png", dpi=200)  # Man kan lägga den i tex assets ../assets/cos(x).png
```

## Subplots för att rita fler grafer 

```python
# MATLAB style
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)  #(number of rows, number of colums, panel number)  # Imatlab räknar man från 1 inte från 0
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")


plt.subplot(1,2,2)
plt.plot(x, f(x))


fig.suptitle("Trigonometric functions", fontweight = "bold")
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,5))  # 1,2 om du tar 1,3 så får du tre grafer istället

ax[0].plot(x, f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)");
```
