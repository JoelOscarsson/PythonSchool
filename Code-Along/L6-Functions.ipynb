## Functions
Basic functions

```python
## Define a function called cuber, it takes in 1 parameter

def cube(number):       
    return number**3

## Vi anropar/call funktionen cube using 1 argument
print(f"{cube(3)}: {cube(3)}")

cube5 = cube(5)
print(cube5)

cubes = [cube(x) for x in range(10)]
print(cubes)
```

```python
import matplotlib.pyplot as plt

plt.plot(cube5);
```
